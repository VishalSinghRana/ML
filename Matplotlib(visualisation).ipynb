## Matplotlib

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.

Some of the major Pros of Matplotlib are:

- Generally easy to get started for simple plots
- Support for custom labels and texts.
- Great control of every element in a  figure
- High quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt

%matplotlib inline 
#%matplotlib inline helps to display the plot without mentioning the plt.show().
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
x
```

```python
y=np.arange(11,21)
```

```python
# scatter plot

plt.scatter(x,y,s=100,c="r",edgecolors="g")
# c stands for color here.
# s is the markersize
# edgecolors gives a color to the edges of the marker.
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Scatter plot")
# xlabel,ylabel and title labels the graph.

# if we wish to save the below graph we can do that by the below code.
# plt.savefig("scatter.png") 
# uncomment the above code and try out yourself.
```

```python
y = x ** 3 
```

```python
# plot

plt.plot(x,y)
# gives a line plot.
```

```python
plt.plot(x,y,"r")
# "r" is the color of the line(red in the case) we can modify it to produce different colors and styles.
# "r--","ro--","r*--","--","-","ro" , etc
```

```python
plt.plot(x,y,"go--")
# we have other arguments like linestyle='dashed',linewidth=2, markersize=12 to modify the line and make the graph more
# attractive
```

```python
plt.plot(x,y,"k*--",linestyle='dashed',linewidth=2, markersize=12)
# k is code for black color
```

### Subplot

```python
# subplot basically means dividing our canvas into a no of small canvas(canvas here is the area in which a plot is shown. 
```

```python
plt.subplot(2,2,1)
# syntax = plt.subplot(nrows,ncols,plotnumber)
plt.plot(x,y,"r*-")
plt.subplot(2,2,2)
plt.plot(x,y,"ro--")
plt.subplot(2,2,3)
plt.plot(x,y,"r--")
plt.subplot(2,2,4)
plt.plot(x,y,"r-")

# in a single plot area we have shown small small 4 plots
```

```python
# making a graph from a equation
# x **2 + 2x
y = x**2 +2 * x
plt.plot(x,y,"--")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Equation graph")
```

```python
np.pi
```

```python
# making a sine wave 
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Sine wave")
```

```python
# plotting cos wave.
y = np.cos(x)
plt.plot(x,y)
plt.title("Cos wave")
plt.xlabel("X axis")
plt.ylabel("Yaxis")
```

```python
# Bar plot

x=[1,6,3]
y=[2,12,6]

x2 =[5,2,7]
y2 =[10,4,14]
plt.bar(x,y)
plt.bar(x2,y2,color="r")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Bar plot")
```

## Histogram

```python
# histogram basically tells about the number of observation between a given interval.
a = np.array([22,33,21,23,27,33,39,34,34.5,41,43,43,44,45,44,43,41,50,20,51])
plt.hist(a)
plt.xlabel("Interval")
plt.ylabel("Observation")
plt.title("Histogram")
```

## Box plot

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# normal(0 is the mean, std will range between 1and 4 not including 4 ,100 is the number of observation it will give)
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()

# making vert=False make the box plots horizontal and patch_artist=False will remove the color
```

```python
data
```

```python
plt.boxplot(data,vert=False,patch_artist=False)
plt.show()
```

## Pie chart

```python
labels = "Fat","Carbs","Sugar","Protein","Salt"
qty= ["70","260","90","50","6"]
colors = ["indigo","Blue","green","yellow","Orange"]
explode = [0,0,0,0.5,0]
plt.pie(qty,explode=explode,labels=labels,colors=colors,autopct="%1.f%%",shadow=True)
plt.axis("equal")
plt.show()
```
