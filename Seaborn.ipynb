# Seaborn

- Distplot,joinplot and pairplot very helpful for numerical data.

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
# tips is a dataset that tells us about the bill of a customer,sex,time of having food adn the tip that customer leaves.
```

```python
df.head()
```

# Correlation

- Correlation basically means what is the relation between the features.

- A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. It is very important in Feature Selection

```python
df.corr()
# correlation ranges between -1 to 1. 
# if 1 is the relation its a strictly increasing relation(as 1 feature increases 2nd feature also increases).
# -1 means decreasin relation (as 1 feature increases the 2nd decreases)
```

```python
# heatmap
sns.heatmap(df.corr())
# darker the color lesser the correlation.
```

# Joinplot

- A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="hex")
# kind can be "scatter" , "reg" , "resid" , "kde" , "hex",Kind of plot to draw,its optional.
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="resid")
# x axis has a histogram of tip and y axis has histogram of total_bill.
# the curve line that you see for total_bill histogram is kde(kernel density estimation)
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind="reg")
# histogram displayed with respective kde
# reg is regression,gives us the regression line.
```

# Pair plot

- A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
# the number of pairplot you get depends on the number of feature you use.
# we get the plots equal to the square of the feature we use.i.e 2 feature will result in 4 plot.
# 3 will result in 9 4 in 16 and so on.
sns.pairplot(df)

# and this plot is also for numerical valued features only same as previos plots. we wont see categorical features here.
```

```python
sns.pairplot(df,hue="smoker")
# with hue we can select a categorical value to make the points the plot more differentiable.
```

# Dist plot

- As the name suggests stands for distribution plot and helps us to find the distribution of features.

```python
sns.distplot(df["tip"])
```

```python
sns.distplot(df["total_bill"])
# gives us a histogram with the percantage of values.Also gives us the kde.
# by default kde is True, we can change it.
```

```python
sns.distplot(df["size"],kde=False)
# size means the number of people attending the meal.
```

# Categorical plots

- Seaborn also helps us perform analysis on categorical data using boxplot,violinplot,countplot and barplot.

```python
# countplot

sns.countplot("sex",data=df)

# counts the number of male and female in df.
```

```python
sns.countplot(x="sex",y="smoker",data=df)

# in count plot you can only pass values for either x or y, and not for both.
```

```python
sns.countplot(x="sex",data=df)
```

```python
sns.countplot(y="sex",data=df)
```

```python
# bar plot

sns.barplot(x="sex",y="smoker",data=df)

# in bar plot out of x and y one has to be numeric. both cannot be categorical.
```

```python
sns.barplot(x="tip",y="sex",data=df)
```

```python
sns.barplot(x="sex",y="tip",data=df,orient="v")
# orient helps to display the plot vertical ("v") or horizontal("h")
```

```python
# boxplot

sns.boxplot("smoker","tip",data=df,orient="v")
```

```python
sns.boxplot("tip","smoker",data=df,orient="h")
```

```python
sns.boxplot("day","tip",data=df,palette="rainbow")
# palette is like giving a color scheme to the box plots.
```

```python
# box plot can also be used for numerical features and not just categorical feartures
sns.boxplot(data=df,orient="v")
```

```python
# making a box plot based on some other category.
# It tells us about smoker and non smoker visiting the hotel from thur to sunday.
# hue can be changed and be made anything

sns.boxplot(x="total_bill",y="day",hue="smoker",data=df)
```

```python
sns.boxplot(x="total_bill",y="sex",hue="smoker",data=df)

# the boxplot tells about the bill differnce between smoker and non smoker male and females.
```

## Violin plot 

- Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot


```python
sns.violinplot("tip","sex",data=df)

# violin plot giving the kde(kernel density estimation) and the box plot along with the distribution of tip.
# the black solid figure in the centre if violin plot is the box plot
```
