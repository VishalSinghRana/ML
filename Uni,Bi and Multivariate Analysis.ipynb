### Univariate,Bivariate and Multivariate Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
# reading the iris dataset.
```

```python
df.head()
# the dataset has 4 features sepal_length,sepal_width,petal_length,petal_width.
```

```python
# if we want to know the total number of rows and column in the dataframe we can use the shape attribute.

df.shape
```

```python
# we have 150 rows and 5 columns
# the species column is our dependent variable it has 3 species setosa,versicolor and virginica
```

### Univariate Analysis

```python
df_setosa=df.loc[df["species"]=="setosa"]
# here we are making a dataframe only for the setosa species.
```

```python
df_setosa
```

```python
# same we will be doing for viginica and versicolor
```

```python
df_versicolor
```

```python
df_virginica = df.loc[df["species"]=="virginica"]
```

```python
df_virginica
```

```python
plt.plot(df_setosa["sepal_length"],np.zeros_like(df_setosa["sepal_length"]),"o")

# here zeros_like() returns an array of zeros with the same shape and type as a given array.
# df_setosa["sepal_length"] is the array that we have given it.
# we want the y-axis to be all of zeros and hence zeros_like is used.
```

```python
plt.plot(df_setosa["sepal_length"],np.zeros_like(df_setosa["sepal_length"]),"o")
plt.plot(df_virginica["sepal_length"],np.zeros_like(df_virginica["sepal_length"]),"o")
plt.plot(df_versicolor["sepal_length"],np.zeros_like(df_versicolor["sepal_length"]),"o")
# showing all the three plots on the same plot.

plt.xlabel("Sepal_Length")
```

### Bivariate Analysis

```python
# will make use of seaborn library
```

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width").add_legend()



#FacetGrid:- Multi-plot grid for plotting conditional relationships.
# here we are creating a scatter plot with petal_length on x axis and sepal_width on y axis.
# hue parameter basically distinguish them based on the given parameter.
```

We can see that the petal_length of setosa flowers lies between 1 and 2,where as in the case of versicolor and virginica 
there's a bit of overlapping.

### Multivariate Analysis

```python
sns.pairplot(df,hue="species",size=2)
```

Here every feature is plotted with every other feature.Pairplot is useful when we have 3-10 features in our data.
The number of plots that we get is square of the number of features we have.



From the plots we can see that petal_length and petal_width are the most adequate feature to work with because we are able
to separate the species properly with the help of petal_length and petal_width.
