```python
df1=pd.DataFrame(lst)
df1
```

```python
df1.dropna(axis = 0) # dropna() drops the whole row or column if a NaN value is present in it.
# axis = 0 means drop the rows
```

```python
df1.dropna(axis=1) # drops the column where NaN values are present.
# here only the index column is where theres no NaN value present.
```

# Reading json to csv

- json data is basically a collection of key value pairs

```python
import pandas as pd
Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df = pd.read_json(Data)

# read_json helps to read the json file
```

```python
df
```

```python
df.to_csv("employee.csv")
# the dataframe has been saved to employee.csv
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data',header=None)
# if we wont mention header the first entry is taken as column name.
# try not mentioning header and check out the result.
```

```python
df
```

```python
# try to save wine data also as a csv
df.to_csv("wine.csv")

# stored to wine.csv
```

```python
# converting a json file to different json format
df.to_json(orient="index")

# here "0":{"0":1,"1":14.23,"2":1.71,"3":2.43,"4":15.6,"5":127,"6":2.8,"7":3.06,"8":0.28,"9":2.29,
# "10":5.64,"11":1.04,"12":3.92,"13":1065} this means the 0th row of the dataframe has the following data.
# change the orient to records
```

```python
df.to_json(orient="records")
# see the difference here the row number isnt mentioned as the key ,here the first entry in curly braces is
# the value of first row for all the 13-14 columns 
# the 0th column of 0th row(1st entry for 1st column ) is 1 ,for 1st is 14.23 and so on .
# you can check with the table
# with orient = record the data is stored row wise like all the values of a first row is stored
# than the second row then third and so on
```
