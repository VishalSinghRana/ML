```python
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
# species is our dependent variable.
```

```python
sns.lmplot(x="sepal_length",y="petal_length",data=df,col="species",col_wrap=3,hue="species")
# col_wrap will show 3 column per row.
```

```python
sns.scatterplot(x="petal_length",y="sepal_width",data=df)
sns.despine(top="True",left="True",right="True",bottom="True")
sns.set_style("darkgrid")
# set_style makes the grid.
# despine() removes the boundary of the plot.
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="petal_length",style="species")
```
